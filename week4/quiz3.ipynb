```python
import urllib.request
import urllib.parse

API = "https://search.naver.com/search.naver"
values = {
    "sm" : "top_hty",
    "fbm" : "0",
    "ie" : "utf8",
    "query" : "권성중"
}
params = urllib.parse.urlencode(values)
url = API+"?"+params
print("url = ", url)

data = urllib.request.urlopen(url).read()
text = data.decode("utf-8")
print(text)
```
